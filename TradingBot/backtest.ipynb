```python
import pandas as pd
df = pd.read_csv('5m.csv')

df.head()
```

```python
df['unix_timestamp']=(pd.to_datetime(df['unix_timestamp'],unit='ms')) 
df.head()
```

```python
df = df.drop(['volume','close_time', 'quote_asset_volume','number_of_trades','taker_buy_base_asset_vol','taker_buy_quote_asset_vol','ignore'], axis = 1)
```

```python
df['year'] = df['unix_timestamp'].dt.year
df['month'] = df['unix_timestamp'].dt.month
df['day'] = df['unix_timestamp'].dt.day
df['hour'] = df['unix_timestamp'].dt.hour
df['minute'] = df['unix_timestamp'].dt.minute

df.head()
```

```python
df.iloc[0,1]
```

```python
orderid = 1
order_status = 0 # 0 = open, 1 = bought, 2 = sold, 3 = cancelled
investment = 1000
slots = 20
slots_used = 0
open_buys = 0
open_sells = 0
roi = 0.05
fees = 0.001
order_size = investment/slots
profit = 0
order_placed_at = df.iloc[0,0]
order_bought_at = ''
order_closed_at = ''
order_cancelled_at = ''
order_life = 0
buy_price = df.iloc[0,2]
units = order_size * (1-fees)/buy_price
sell_price = order_size * (1+roi) / (units * (1-fees))
order = [orderid,order_status, order_size, buy_price, sell_price, profit, order_placed_at, order_bought_at, order_closed_at, order_cancelled_at, order_life]

order_data = [order]

print(order_data)
```
